```python
import pandas as pd
import numpy as np
```

```python
# Muat dataset Titanic
URL = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(URL)
df
```

```python
#Explorasi Awal
print(df.shape, df.columns.tolist())
print(df.isnull().sum())
```

```python
#Filtering - penumpang wanita kelas 1
wk1 = df[(df['Sex']=='female') & (df['Pclass']==1)]
print(wk1['Survived'].mean(), wk1['Age'].mean())
```

```python
# Groupby - tingkat keselamatan per kelas
hasil = df.groupby('Pclass').agg(
    total = ('PassengerId', 'count'),
    selemat = ('Survived', 'mean')
)
print(hasil)
```

```python
# PERTANYAAN KUIS
# Q1 : BERAPA JUMLAH BARIS DAN KOLOM DATASET TITANIC ?
print(f"Jumlah baris dan kolom: {df.shape}\n")
print(f"Kolom-kolom pada dataset: \n{df.columns.tolist()}\n")
```

```python
# Q2 : KOLOM MANA YANG MEMILIKI NILAI KOSONG (MISSING)?
print("Missing value per kolom:\n", df.isnull().sum())
print("\nPersentase missing value (%):\n", df.isnull().mean() * 100)
```

```python
# Q4 : FILTERING PENUMPANG WANITA DARI KELAS 1 (Pclass=1). BERAPA JUMLAHNYA DAN BERAPA RATA RATA USIANYA?
# wk1 = df[(df['Sex']=='female') & df['Pclass']==1]
print(f"Jumlah penumpang wanita di Kelas 1 : {wk1.shape[0]} orang")
print(f"Rata-rata usia penumpang kelas 1 adalah : {wk1['Age'].mean()} tahun")
```
